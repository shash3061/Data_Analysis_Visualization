### Checking for null values

```python
data.isnull().sum()
```

### Dropping the unnecessary columns

```python
data.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

### Filling the null values

```python
cols = [
    'ShortPassing',
    'Dribbling',
    'Volleys',
    'FKAccuracy',
    'Curve',
    'LongPassing',
    'BallControl',
    'HeadingAccuracy',
    'Finishing',
    'Crossing',
]
```

```python
weights_data = data.copy()
weights_data = weights_data["Weight"].str[:-3]
weights_data = pd.DataFrame(weights_data)
weights_data = weights_data.dropna()
weights = np.array(weights_data)
weights = weights.astype(int)
```

```python
weight_mean = str(int(weights.mean())) + "lbs"
```

```python
for col in cols:
    data[col].fillna(int(data[col].mean()), inplace = True)
```

```python
data['Weight'].isnull().sum()
```

```python
data['Weight'].fillna(weight_mean, inplace = True)
```

```python
data['Weight'].isnull().sum()
```

```python
data['Club'].isnull().sum()
```

```python
data.drop([
    'Contract Valid Until',
    'Height',
    'Loaned From',
    'Jersey Number',
    'Joined',
    'ID',
    'Photo',
    'Flag',
    'Club Logo',
    'LS',
    'ST',
    'RS',
    'LW',
    'LF',
    'CF',
    'RF',
    'RW',
    'LAM',
    'CAM',
    'RAM',
    'LM',
    'LCM',
    'CM',
    'RCM',
    'RM',
    'LWB',
    'LDM',
    'CDM',
    'RDM',
    'RWB',
    'LB',
    'LCB',
    'CB',
    'RCB',
    'RB',
], axis = 1, inplace = True)
```

```python
data['Body Type'].fillna('Normal', inplace = True)
data['Position'].fillna('ST', inplace = True)
data['Club'].fillna('No Club', inplace = True)
data['Work Rate'].fillna('Medium/ Medium', inplace = True)
data['Skill Moves'].fillna(data['Skill Moves'].median(), inplace = True)
data['Weak Foot'].fillna(data['Skill Moves'].median(), inplace = True)
data['Preferred Foot'].fillna('Right', inplace = True)
data['International Reputation'].fillna(1, inplace = True)
data['Wage'].fillna('€200K', inplace = True)
```
